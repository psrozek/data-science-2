## Binary Images

#### cv2.threshold(img, threshold, max, cv2.THRESH_BINARY)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('zebra.jpg')
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img, cmap='gray')
plt.show()
```

```python
thresh = 150
max_val = 255
th, dst = cv2.threshold(img, thresh, max_val, cv2.THRESH_BINARY)
```

```python
plt.imshow(dst, cmap='gray')
plt.show()
```

## Bitwise Operations

```python
board1 = cv2.imread('board.png')
board2 = cv2.imread('board2.png')

board1 = cv2.cvtColor(board1, cv2.COLOR_BGR2GRAY)
board2 = cv2.cvtColor(board2, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(board1, cmap='gray')
plt.show()
```

```python
plt.imshow(board2, cmap='gray')
plt.show()
```

```python
thresh = 150
max_val = 255
```

```python
th, img1 = cv2.threshold(board1, thresh, max_val, cv2.THRESH_BINARY)
th, img2 = cv2.threshold(board2, thresh, max_val, cv2.THRESH_BINARY)
```

```python
plt.imshow(img1, cmap='gray')
plt.show()
```

```python
plt.imshow(img2, cmap='gray')
plt.show()
```

```python
moved = cv2.bitwise_xor(img1, img2)  # XOR used to detect moving pieces between board1 and board2
```

```python
plt.imshow(moved, cmap='gray')
plt.show()
```

## Masking
#### np.where(mask, image, 0)

```python
recording_img = cv2.imread('recording.jpg')
recording_img = cv2.cvtColor(recording_img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(recording_img, cmap='gray')
plt.show()
```

```python
thresh = 150
max_val = 255
```

```python
th, recording_img2 = cv2.threshold(recording_img, thresh, max_val, cv2.THRESH_BINARY)
```

```python
plt.imshow(recording_img2, cmap='gray')
plt.show()
```

```python
plt.imshow(img, cmap='gray')
plt.show()
```

```python
img.shape  #shape of zebras
```

```python
recording_img2.shape  # shape of recording
```

```python
res_img = cv2.resize(img, (1280,800), interpolation=cv2.INTER_LINEAR)
res_recording = cv2.resize(recording_img2, (1280,800), interpolation=cv2.INTER_LINEAR)
```

```python
plt.imshow(res_img, cmap='gray')
plt.show()
```

```python
plt.imshow(res_recording, cmap='gray')
plt.show()
```

```python
res_img.shape  #shape of zebras after resizing
```

```python
res_recording.shape  # shape of recording after resizing
```

```python
result = np.where(res_recording, res_img, 0)  # masking
```

```python
plt.imshow(result, cmap='gray')
plt.show()
```
