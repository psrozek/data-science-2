## Histograms

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('river_scene.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img, cmap='gray')
plt.show()
```

### plotting histogram with matplotlib

#### histogram of grayscale image

```python
ax = plt.hist(img.ravel(), bins=256)
plt.show()
```

```python
plt.hist(img.ravel())  # bins=10 by default
plt.show()
```

#### histogram of BGR image

```python
img2 = cv2.imread('sunflower.jpg')
```

```python
plt.imshow(img2[:,:,::-1])
plt.show()
```

```python
r = img2[:, :, 2]
g = img2[:, :, 1]
b = img2[:, :, 0]
```

```python
plt.hist(r.ravel(), bins=256, color='Red', alpha=0.5)  # alpha is for transparency
plt.hist(g.ravel(), bins=256, color='Green', alpha=0.5)
plt.hist(b.ravel(), bins=256, color='Blue', alpha=0.5)
plt.show()
```

### plotting histogram with opencv
#### cv2.calcHist([images], [channels], mask, [histSize], [ranges])
#### ROI selector: x,y,w,h = cv2.selectROI(img, fromCenter=0)

```python
img3 = cv2.imread('flowers.jpeg')
```

```python
x,y,w,h = cv2.selectROI(img3, fromCenter=0)  # ROI selector appears
```

```python
mask1 = np.zeros(img3.shape[:2], dtype="uint8")  # create array for mask
```

```python
mask1[int(y):int(y+h), int(x):int(x+w)] = 255  # make the selected area white
```

```python
plt.imshow(mask1, cmap='gray')
plt.show()
```

```python
cv2.imwrite('mask1.png', mask1)  # save the mask file
```

```python
hist1 = cv2.calcHist([img3], channels=[0,1,2], mask=mask1, histSize=[10,10,10], ranges=[0, 255, 0, 255, 0, 255])
```

```python
print ("4D histogram shape: %s, with %d values" %(hist1.shape, hist1.flatten().shape[0]))
```
